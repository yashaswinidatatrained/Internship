### 1. Scrape the details of most viewed videos on YouTube from Wikipedia.Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos You need to find following details:

A)Rank
B) Name
C) Artist
D) Upload date
E) Views

```python
#opening the most viewed youtube videos page on automated chrome browser
driver = webdriver.Chrome()
driver.get('https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos')
time.sleep(2)
```

```python
rank = []
name = []
artist = []
upload_date = []
views = []

#to scrape Rank

Rank = driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[1]')
for a in Rank:
    rank.append(a.text)
    
#to scrape Name

Name = driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[2]')
for b in Name:
    name.append(b.text)


#to scrape Artist

Artist = driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[3]')
for c in Artist:
    artist.append(c.text)
    
#to scrape date

Date = driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[5]')
for d in Date:
    upload_date.append(d.text)
    
    
#to scrape views

view = driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[4]')
for e in view:
    views.append(e.text)
```

```python
print(len(rank), len(name), len(artist), len(upload_date), len(views))
```

```python
Youtube = pd.DataFrame({'Rank':rank, 'Video Name': name, 'Artist': artist, 'Views(in Billions)': views, 'Upload Date': upload_date})
Youtube
```

```python
Youtube.to_csv = ("ybmostviewedvideos.csv")
```

```python
driver.close()
```

### 4. Scrape the details of trending repositories on Github.com. Url = https://github.com/ 

You have to find the following details: 

A) Repository title 
B) Repository description 
C) Contributors count 
D) Language used 

Note: - From the home page you have to click on the trending option from Explore menu through code. 

```python
#to open github main page
driver = webdriver.Chrome()
driver.get('https://github.com/')
time.sleep(2)
```

```python
# click on menu  button 
menu = driver.find_element(By.XPATH, "/html/body/div[1]/div[1]/header/div/div[1]/div[2]/button/span/span")
menu.click()
```

```python
#selct open source
open_source = driver.find_element(By.XPATH, "/html/body/div[1]/div[1]/header/div/div[2]/div/nav/ul/li[3]/button")
open_source.click()
```

```python
#click on trending pages
trending = driver.find_element(By.XPATH,'/html/body/div[1]/div[1]/header/div/div[2]/div/nav/ul/li[3]/div/div[3]/ul/li[2]/a')
trending.click()
```

```python
title = []
description = []
con_count = []
language = []

#to scrape title 
r_title=driver.find_elements(By.XPATH,'//a[@class="Link"]')
for a in r_title:
    title.append(a.text)
    
#to scrape url of all repository
r_URL = []
url_tags=driver.find_elements(By.XPATH,'//a[@class="Link"]')
for t in url_tags:
    r_URL.append(t.get_attribute('href'))
    
#to scrape desscription, contributor count, language used
for t in r_URL:
    driver.get(t)
    time.sleep(5)
    
    try:
        desc=driver.find_element(By.XPATH,'//p[@class="f4 mb-3 "]')
        description.append(desc.text)
    except NoSuchElementException:
        description.append('-----')
        
        
    try:
        count=driver.find_element(By.XPATH,'//*[contains(text(),"   Contributors")]/span')
        con_count.append(count.text)
    except NoSuchElementException:
        con_count.append('-----')
    

    try:
        lang =  driver.find_element(By.XPATH,'//li[@class="d-inline"]')
        language.append(lang.text.replace('\n','  '))
    except NoSuchElementException:
        language.append('-----')
```

```python
print(len(title),len(description),len(con_count),len(language ))
```

```python
github = pd.DataFrame({'Repository title': title,'Repository description' :description,'Contributors count':con_count,'Language used':language})
github
```

```python
github.to_csv('ws4trendingrepo.csv')
```

### 7. Scrape the details most watched tv series of all time from imdb.com. Url = https://www.imdb.com/list/ls095964455/ 

You have to find the following details: 

A) Name 
B) Year span 
C) Genre 
D) Run time 
E) Ratings 
F) Votes

```python
#open imdb page
driver = webdriver.Chrome()
driver.get('https://www.imdb.com/list/ls095964455/')
time.sleep(2)
```

```python
#to scrape most watched series details 
name =[]   
year = []   
genre = []  
run = []  
ratings = []   
votes = []   

video = driver.find_elements(By.XPATH,'//div[@class="lister-item-content"]/h3/a')
for a in video:
    name.append(a.text)
    
span = driver.find_elements(By.XPATH,'//div[@class="lister-item-content"]/h3/span[2]')
for b in span:
    year.append(b.text)
    
serie = driver.find_elements(By.XPATH,'//div[@class="lister-item-content"]/p[1]/span[5]')
for c in serie:
    genre.append(c.text)
    
time = driver.find_elements(By.XPATH,'//div[@class="lister-item-content"]/p[1]/span[3]')
for d in time:
    run.append(d.text)
    
rate = driver.find_elements(By.XPATH,'//div[@class="lister-item-content"]/div[1]/div[1]/span[2]')
for e in rate:
    ratings.append(e.text)
    
voting = driver.find_elements(By.XPATH,'//div[@class="lister-item-content"]/p[4]/span[2]')
for f in voting:
    votes.append(f.text)
    
print(len(name), len(year), len(genre), len(run),len(ratings),len(votes))
    
```

```python
#create dataframe
imdb = pd.DataFrame({'Name':name,'Year Span': year, 'Genre': genre, 'Run time': run, 'Ratings': ratings,'Votes': votes})
imdb
```

```python
imdb.to_csv('ws4imdb.csv')
```

### 8. Details of Datasets from UCI machine learning repositories. Url = https://archive.ics.uci.edu/ 
You have to find the following details: 

A) Dataset name 
B) Data type 
C) Task 
D) Attribute type 
E) No of instances 
F) No of attribute G) Year

 Note: - from the home page you have to go to the Show All Dataset page through code.

```python
#to open archive.ics.uci automated page on browser
driver =webdriver.Chrome()
driver.get('https://archive.ics.uci.edu/')
time.sleep(2)
```

```python
#to click on view all
viewall = driver.find_element(By.XPATH,'//a[@class="btn-primary btn"]')
viewall.click()
```

```python
# to scrap URL of 66 pages
start =0
end = 66

Pr_URL = []

for page in range(start,end):
    url_tags= driver.find_elements(By.XPATH,'//a[@class="link-hover link text-xl font-semibold"]')
    for t in url_tags:
        Pr_URL.append(t.get_attribute('href'))
    next_page = driver.find_element(By.XPATH,'//div[@class="my-4 flex flex-wrap items-center justify-end gap-4"]/div/button[2]')
    next_page.click()
    time.sleep(3)
```

```python
len(Pr_URL)
```

```python
Dname =[]
Dtype = []
task = []       
attritype = []
instances = []
attribute = []
year = []


for t in Pr_URL:
    driver.get(t)
    time.sleep(5)
    
    try:
        name = driver.find_element(By.XPATH,'//h1[@class="text-3xl font-semibold text-primary-content"]')
        Dname.append(name.text)
    except NoSuchElementException:
        Dname.append('-----')
        
    try:
        datatype = driver.find_element(By.XPATH,'//div[@class="grid grid-cols-8 gap-4 md:grid-cols-12"]/div[1]/p')
        Dtype.append(datatype.text)
    except NoSuchElementException:
        Dtype.append('-----')
        
    try:
        dtask = driver.find_element(By.XPATH,'//div[@class="grid grid-cols-8 gap-4 md:grid-cols-12"]/div[3]/p')
        task.append(dtask.text)
    except NoSuchElementException:
        task.append('-----')
        
    try:
        attype = driver.find_element(By.XPATH,'//div[@class="grid grid-cols-8 gap-4 md:grid-cols-12"]/div[4]/p')
        attritype.append(attype.text)
    except NoSuchElementException:
        attritype.append('-----')
               
    try:
        inst = driver.find_element(By.XPATH,'//div[@class="grid grid-cols-8 gap-4 md:grid-cols-12"]/div[5]/p')
        instances.append(inst.text)
    except NoSuchElementException:
        instances.append('-----')
        
    try:
        count = driver.find_element(By.XPATH,'//div[@class="grid grid-cols-8 gap-4 md:grid-cols-12"]/div[6]/p')
        attribute.append(count.text)
    except NoSuchElementException:
        attribute.append('-----')
        
    try:
        date = driver.find_element(By.XPATH,'//div[@class="relative flex flex w-full items-center gap-4 bg-primary p-2"]/div[2]/h2')
        year.append(date.text)
    except NoSuchElementException:
        year.append('-----')
```

```python
print(len(Pr_URL), len(Dname),len(Dtype),len(task),len(attritype),len(instances),len(attribute), len(year))
```

```python
uci = pd.DataFrame({'Dataset name':Dname,'Data type': Dtype,'Task':task,'Attribute type':attritype,'No of instances':instances,
                    'No of attribute': attribute, 'Year': year})
uci
```

```python
uci.to_csv = ('ws4uci.csv')
```
