### 5) Write a python program to scrape mentioned news details from https://www.cnbc.com/world/?region=world and make data frame

i) Headline
ii) Time
iii) News Link

```python
page = requests.get("https://www.cnbc.com/world/?region=world")
page
```

```python
soup = BeautifulSoup(page.content)
soup
```

```python
title = []

for i in soup.find_all('a',class_= "LatestNews-headline"):
    title.append(i.text)

title
```

```python
time = []

for i in soup.find_all('time',class_= "LatestNews-timestamp"):
    time.append(i.text)

time
```

```python
link = []

for i in soup.find_all('a', class_="LatestNews-headline"):
    link.append(i.get('href'))

link
```

```python
print(len(title),len(time),len(link))
```

```python
import pandas as pd
CNBC=pd.DataFrame({"Headline": title,"Time": time, 'Newslink': link})
CNBC
```

#### 7) Write a python program to scrape mentioned details from dineout.co.in and make data frame

i) Restaurant name
ii) Cuisine
iii) Location
iv) Ratings
v) Image URL

```python
page = requests.get("https://www.dineout.co.in/delhi-restaurants/buffet-special")
page
```

```python
soup = BeautifulSoup(page.content)
soup
```

```python
name = []
for i in soup.find_all('a',class_="restnt-name ellipsis"):
    name.append(i.text)
    
name
```

```python
cuisine = []
for i in soup.find_all('span', class_="double-line-ellipsis"):
    cuisine.append(i.text)
    
cuisine
```

```python
rate = []
for i in soup.find_all('div',class_="restnt-rating rating-4"):
    rate.append(i.text)
    
rate
```

```python
loc = []
for i in soup.find_all('div',class_="restnt-loc ellipsis"):
    loc.append(i.text)
    
loc
```

```python
images_url = []

for i in soup.find_all('img', class_="no-img"):
    images_url.append(i['data-src'])

images_url
```

```python
print(len(name),len(loc),len(cuisine),len(rate),len(images_url))
```

```python
DineOut = pd.DataFrame({'Restaurant Name': name,'Cuisine': cuisine,'Location': loc,'Ratings':rate,'Images Url':images_url})
DineOut
```
