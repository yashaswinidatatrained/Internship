### Write a python program which searches all the product under a particular product from www.amazon.in. The product to be searched will be taken as input from user. For e.g. If user input is ‘guitar’. Then search for guitars. 

```python
Pr_name=input("Which product do you need to search?  :")

driver = webdriver.Chrome()
driver.get("https://www.amazon.in/")
time.sleep(2)

# get web element for amazon search bar
guitar = driver.find_element(By.XPATH,'//input[@type="text"]')
#enter the product into search bar
guitar.send_keys(Pr_name)

#to click search button
search = driver.find_element(By.XPATH,"//*[contains(@type,'sub')]")
search.click()
```

### In the above question, now scrape the following details of each product listed in first 3 pages of your search results and save it in a data frame and csv. In case if any product has less than 3 pages in search results then scrape all the products available under that product name. Details to be scraped are: "Brand Name", "Name of the Product", "Price", "Return/Exchange", "Expected Delivery", "Availability" and “Product URL”. In case, if any of the details are missing for any of the product then replace it by “-“. 

```python
# to scrap URL
start =0
end = 3

Pr_URL = []
for page in range(start,end):
    url_tags=driver.find_elements(By.XPATH,'//a[@class="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"]')
    for t in url_tags:
        Pr_URL.append(t.get_attribute('href'))
next_page = driver.find_elements(By.XPATH,'//a[@class="s-pagination-item s-pagination-next s-pagination-button s-pagination-separator"]')
```

```python
len(Pr_URL)
```

```python

Brand =[]
Pr_Name = []
Price = []       
Ret_exc = []
Exp_Del = []
Avail = []

for t in Pr_URL:
    driver.get(t)
    time.sleep(5)
    
    try:
        brand = driver.find_element(By.XPATH,'//div[@class="a-section a-spacing-medium brand-snapshot-flex-row"]/p')
        Brand.append(brand.text)
    except NoSuchElementException:
        Brand.append('-----')
        
    try:
        product = driver.find_element(By.XPATH,'//span[@class="a-size-large product-title-word-break"]')
        Pr_Name.append(product.text)
    except NoSuchElementException:
        Pr_Name.append('-----')
    
    try:
        cost = driver.find_element(By.XPATH,'//span[@class="a-price aok-align-center reinventPricePriceToPayMargin priceToPay"]')
        Price.append(cost.text)
    except NoSuchElementException:
        Price.append('-----')
        
    try:
        return_exc = driver.find_element(By.XPATH,'//div[@id="icon-farm-container"]/div/div/div/div/ol/li[3]')
        Ret_exc.append(return_exc.text)
    except NoSuchElementException:
        Ret_exc.append('-----')

    try:
        delivery = driver.find_element(By.XPATH,'//div[@class="a-spacing-base"]/span/span[1]')
        Exp_Del.append(delivery.text)
    except NoSuchElementException:
        Exp_Del.append('------')

    try:
        avail = driver.find_element(By.XPATH,'//span[@class="a-size-medium a-color-success"]')
        Avail.append(avail.text)
    except NoSuchElementException:
        Avail.append('------')
```

```python
print(len(Brand),len(Pr_Name),len(Price),len(Ret_exc),len(Exp_Del),len(Avail))
```

```python
T_SHIRT = pd.DataFrame({'Brand':Brand,
'Product_Name': Pr_Name,
'Price': Price,
'Return/Exchange': Ret_exc,
'Exppected_Delivery': Exp_Del,
'Availability': Avail,
'Product_Url' : Pr_URL})
T_SHIRT
```

```python
T_SHIRT.to_csv("T_shirt_amazon_ws.csv")
```

```python
Name = []
name = driver.find_elements(By.XPATH,'//div[@class="content"]/div/div/div/div[5]/div/a/a/div[2]/div[1]/span')
for i in name:
    hostel_names = i.text
    Name.append(hostel_names)
    
#to scrape review details    
review_details =[i.text.split('\n') for i in driver.find_elements(By.XPATH,'//div[@class="content"]/div/div/div/div[5]/div/a/a/div[2]/div[2]/div/div[1]')] 
Rating = []    
Review = []

for i in review_details:
    
    try:
        Rating.append(i[0])
        Review.append(i[1])
    except:
        Review.append('-------')
        
        
#to scrape overall_reviews
total_review = []
total = driver.find_elements(By.XPATH,'//div[@class="content"]/div/div/div/div[5]/div/a/a/div[2]/div[2]/div/div[2]')
for j in total:
    count = j.text
    total_review.append(count)
    
    
#to scrape distance
distance = []
km = driver.find_elements(By.XPATH,'//div[@class="content"]/div/div/div/div[5]/div/a/a/div[2]/div[3]/span[2]')
for k in km:
    dist = k.text
    distance.append(dist)
    
    
#to scrape price 
private = [i.text.split('\n') for i in driver.find_elements(By.XPATH,'//div[@class="content"]/div/div/div/div[5]/div/a/a/div[2]/div[5]')]


```

```python
#to scrape Urls

pr_url = []
url_tags=driver.find_elements(By.XPATH,'//div[@class="content"]/div/div/div/div[5]/div/a/a')
for a in url_tags:
    pr_url.append(a.get_attribute('href'))
```

```python
HD = []
Price = []
Free_fac = []
Gen_fac = []

for t in pr_url:
    driver.get(t)
    time.sleep(5)
    
    try:
        description = driver.find_element(By.XPATH,'//div[@class="content collapse-content"]')
        HD.append(description.text)
    except NoSuchElementException:
        HD.append('-')
    try:
        cost = driver.find_element(By.XPATH,'//div[@class="availability-table"]/div[2]/div[1]/div[2]/div[1]/div[1]/div[1]/div[2]/div[1]')
        Price.append(cost.text)
    except NoSuchElementException:
        Price.append('-') 
    try:
        free = driver.find_element(By.XPATH,'//div[@class="facilities-container"]/ul/li[1]/ul')
        Free_fac.append(free.text)
    except NoSuchElementException:
        Free_fac.append('-')
    try:
        gen = driver.find_element(By.XPATH,'//div[@class="facilities-container"]/ul/li[2]/ul')
        Gen_fac.append(gen.text)
    except NoSuchElementException:
        Gen_fac.append('-')        
        
        
```
